```python
#Importar la liberia de pandas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Leer el archivo csv
df = pd.read_csv('titanic.csv')
#Llamar a la variable df
df
```

```python
#1.Cuantos pasajeros eran de primera clase

# Filtrar el DataFrame para encontrar las filas
a =df[df['Pclass']==1]
x = a['Age'] #asignamos a x el campo Edad
y = a['Pclass'] #asignamos a y el campo de clase

plt.bar(x,y) #Grafica de barras
plt.title('Pasajeros de Clase A') #Poner titulo
plt.xlabel('Edad') #Asignar a x el nombre edad
plt.ylabel('Clase') #Asignar a xÂ¿y el nombre Clase
plt.show()  #Ejecutar
```

```python
#2.Cuantos mujeres de primera clase habÃ­a

# Filtrar el DataFrame para encontrar las filas
a=df[df['Pclass']==1]
#Filtra por el sexo de mujer
b=a[a['Sex']=='female']
x = b['Age']  #asignamos a x el campo Edad
y = b['Pclass'] #asignamos a y el campo Clase

plt.bar(x,y) #Grafica de barras
plt.title('Pasajeros de Clase A Mujeres') #Poner titulo
plt.xlabel('Edad') #Asignar a x el nombre edad
plt.ylabel('Clase') #Asignar a y el nombre CLase
plt.show()  #Ejecutar
```

```python
#3.Cuantos hombres de primera y tercera clase habÃ­a en el barco



b=df[(df['Pclass']==1) | (df['Pclass']==3)] #Filtra por sexo de hombre la variable c
c=b[b['Sex']=='male']
x = c['Age'] #asignamos a x el campo Edad
y = c['Pclass'] #asignamos a y el campo Clase

plt.bar(x,y) #Grafica de barras
plt.title('Pasajeros de Clase A y C. Hombres') #Poner titulo
plt.xlabel('Edad')  #Asignar a x el nombre edad
plt.ylabel('Clase')  #Asignar a y el nombre clase
plt.show()  #Ejecutar
```

```python
#4.Muestra el total de la informaciÃ³n de los pasajeros mayores de edad


a=df[df['Age'] >= 18]
x = a['Age'] #asignamos a x el campo edad
y = a['Pclass'] #asignamos a y el campo Clase

plt.bar(x,y) #Grafica de barras
plt.title('Pasajeros mayores de edad') #Poner titulo
plt.xlabel('Edad') #Asignar a x el nombre edad
plt.ylabel('Clase') #Asignar a y el nombre Clase
plt.show()  #Ejecutar
```

```python
#5.Muestra la cantidad de hombres  y mujeres que iban en el barcao

grupoSexo = df.groupby('Sex')['Sex'].count().plot(kind='bar')
```

```python
#6.Muestra la cantidad de hombres mayores a 20

a=df[df['Age'] >= 20]
b=a[a['Sex']=='male'] #Filtra la los mayores de 20 mediante hombres
x = b['Name']  #asignamos a x el campo nombre
y = b['Age']  #asignamos a y el campo EDad

plt.bar(x,y) #Grafica de barras
plt.title('Pasajeros hombres mayores de 20') #Poner titulo
plt.xlabel('Nombre')  #Asignar a x el nombre Nombre
plt.ylabel('Edad')  #Asignar a y el nombre edad
plt.show()  #Ejecutar
```

```python
#7.Muestra la cantidad de pasajeros de 22, 34 y 44 aÃ±os que iban en el barco


a=df[(df['Age'] == 22) | (df['Age'] == 34) | (df['Age'] == 44) ]
x = a['Age'] #asignamos a x el campo edad
y = a['Name'] #asignamos a y el campo nombre

plt.bar(x,y) #Grafica de barras
plt.title('Pasajeros de 22, 34 y 44 abordo') #Poner titulo
plt.xlabel('Edad') #Asignar a x el nombre Edad
plt.ylabel('Nombre') #Asignar a y el nombre Nombre
plt.show()  #Ejecutar
```
